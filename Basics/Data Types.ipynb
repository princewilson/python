# Data Types in Python

Python has the following data types *built-in* by default:

- Text Type: `str`
- Numeric Types: `int`, `float`, `complex`
- Sequence Types: `list`, `tuple`,`range`
- Mapping Type: `dict`
- Set Types: `set`,`frozenset`
- Boolean Type: `bool`
- Binary Types: `bytes`,`bytearray`,`memoryview`

To know which variable is of which type, use `type(variable_name)` method

## Text Type: `str`

- Anything placed inside single or double quotes is said to be of `str` type
- `str` means string

```python
a = 'Prince Wilson'
b = '123'
c = 'Prince1805'
print(type(a))
print(type(b))
print(type(c))
```

## Numeric Types: `int`,`float`,`complex`

### `int` type:

- Any number, positive or negative, not having a fractional part comes under the domain of 'int' type

```python
a = 10
b = 1234567
c = -11
print(type(a))
print(type(b))
print(type(c))
```

### `float` type:

- Any number having a fractional part is considered to be of type `float`

```python
a = 12.25
b = 10.0
c = 0.0
d = -12.5
print(type(a))
print(type(b))
print(type(c))
print(type(d))
```

### `complex` type:

- Any number in the form a + bj, where a and b are real numbers and j is defined as square root of -1, is said to be of `complex` type.

```python
a = 3 + 4j
b = 3.5 + 4.5j
c = 35 + 5.8j
d = 3.5 + 5j
print(type(a))
print(type(b))
print(type(c))
print(type(d))
```

## Sequence Types: `list`,`tuple`,`range`

### `list` type:

- `list` is a data structure which stores different types of mutable data each seperated by comma.
- Syntax of `list` is:

`variable-name = [item 0, item 1, item 2,...]`

```python
names = ['Prince','Ronny','Harshavardhan']
print(type(names))
print(names)
```

```python
different = [1, 2, 3, 4.5, 'my', 'name', "Prince"]
print(type(different))
print(different)
```

## `tuple` type:

- `tuple` is a data structure which stored different type of data seperated by comma, but are immutable.
- Syntax of `tuple` is:

`variable_name = (item 0, item 1, item 2,...)`

```python
names = ('Prince','Ronny','Harshavardhan')
print(type(names))
print(names)
```

## `range` type:

- The `range()` functions returns a sequence of numbers starting from 0 (*by default*) and increments by 1(*by default*) and stops at a particular number.
