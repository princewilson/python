# Control Flow Statements:

- Control flow statements determine in which flow the program is to be executed.

- While executing a time will come when the compiler will need to take a decision based on the condition. With the help of control flow statements we can decide what decision is to be taken.

- Keywords used here are : `if`, `else` and `elif`

- Basic syntax of any control flow statement is :
    
`if condition:
       block of statements
 elif condition:
       block of statements
 else:
       block of statements`
        
- In the above syntax, the `elif` and `else` statements are optional. 

- The meaning of the syntax is just how it sounds like, i.e if the condition is true then execute the following statements and if not just execute the else statement.

- Pay close attention to the indentation.

## Program 1:

### Write a program to check if the given number is even or odd.

```python
if int(input('Enter a number')) % 2 == 0:
    print('Even!')
else:
    print('Odd!')
```

You can also intialize some variable say a and store int(input('Enter a number')) value in a to make it more readable.

## Program 2:

### Write a program to find the grade of a student when he inputs his percentage.

```python
percent = int(input('Enter your percentage: '))
if percent >= 90 and percent <= 100:
    print('Your grade is : A+')
elif percent >= 80:
    print('Your grade is : A')
elif percent >= 70:
    print('Your grade is : B')
elif percent >= 60:
    print('Your grade is : C')
elif percent >= 50:
    print('Your grade is : D')
elif percent < 50 and percent >= 0:
    print('You\'ve utterly failed!!!')
else:
    print('Invalid input!\nTry again!')
```

# Loops:

- Loops are used for repeating the same statement as long as the given condition is satisfied.

- We have two types of loops in Python:
    - `for` loop
    - `while` loop

### `for` loop:

Syntax:

<p style ="font-family : 'Courier New'; font-size : 20px"><b>for</b> variable_name <b>in</b> iterable_name:<br>
        &nbsp;&nbsp;&nbsp;&nbsp;Block of Statements with equal indentation</p>


```python
for i in range(1, 10):
    print(i, end = ' ')
```

```python
for i in range(10):
    print(i, end = ' ')
```

```python
for i in range(1, 11, 2):
    print(i, end = ' ')
```

```python
for i in reversed(range(0, 11)):
    print(i, end = ' ')
```

```python
names = ['Prince', 'Ronny', 'Sowjanya']
for i in names:
    print('Your name is {}'.format(i))
```

### `while` loop:

<h3>Syntax:</h3>

<p style = "font-family : 'Courier New'; font-size : 20px;"><b>while</b> condition:<br>
    &nbsp;&nbsp;&nbsp;&nbsp;Block of statements with equal indentation

### Write a program to print the reverse of a given number:

```python
n = int(input('Enter a number: '))
while n > 0:
    rem = n % 10
    print(rem, end = '')
    n = n // 10
```
