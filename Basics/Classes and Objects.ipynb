## Accessing a Class:

<p style= "font-size: 15px; font-family: 'Courier New'" >- There are two ways of accessing the variables and methods of a class: <i>Attribute Reference</i> and <i>Instantiation</i></p>

### Attribute References:

<p style= "font-size: 15px; font-family: 'Courier New'" >- You can access a variable of a particular class using <b>classname.variablename</b><br>- Whereas to call a function present inside a class you must use: <b>classname.functionname(classname())</b>. The reason for passing an argument with the classname and a paranthesis is because every function in a class is defined with atleast one parameter called <b>self</b> which basically means the class itself.<br>- The function in a class takes the object of it's own class as a parameter and to pass a class object we include paranthesis beside the class name.</p>

```python
class MyClass:
    a = 4
    def show(self):
        print('The value of a is {}'.format(self.a))

print('Value of a is: {}'.format(MyClass.a))
MyClass.a = 5
MyClass.show(MyClass())
```

### Instantiation:

<p style= "font-size: 15px; font-family: 'Courier New'">- Instantiation means to create an object of a class and store it in a variable.<br>- Now, this variable can be used to access the properties of a class.<br>- You can also create new variables for that class by initializing it with the class object in the format: <b>objectname.variablename = value</b><br>- With the below given self-explanatory example, you must have got a basic understanding on class objects.</p>

```python
class MyClass:
    def show(self):
        print('This is the method show.')

x = MyClass()
MyClass.show(x)
MyClass().show()
x.show()
```

### `__init__()` for fast initialization:

<p style= "font-size: 15px; font-family: 'Courier New'">- If you have learned Object Oriented Programming languages like Java or C++, you must have heard about the term <i>Constructor</i>.<br>- __init__() method is similar to a constructor. If it is defined in any class then it is automatically invoked the moment the class is instantiated.<br>- Just like the general use of constructors in other languages, __init__() method is used to initialize the class variables.</p>

```python
class Complex:
    def __init__(self, real, imaginary):
        self.real = real
        self.imaginary = imaginary
    
    def show(self):
        print('The complex number is: {} + i({})'.format(self.real, self.imaginary))

num = Complex(3, 4)
num.show()
```

## Inheritance

<p style = "font-size: 15px; font-family: 'Courier New'">- Like any other object oriented programming language, python also supports the concept of inhertance.<br>- For a quick recap, inheritance is the property of being able to use the methods of another class in the current class which you are working on.<br>- This is extremely useful as it helps save time in avoiding redundant code, without copy pasting all the code in your previous projects into the current one.<br>- The class which is being inherited by another class is known as the Base/Parent/Super class.<br> - The class which is inheriting another class' attributes is known as the Derives/Child/Sub class.<br> - In the below sample program, Money is the Base/Parent/Super class, whereas Happiness is the Derived/Child/Sub class.</p>

```python
class Money:
    def earn_money(self):
        print('You have earned a Million Dollars today!')

class Happiness(Money):
    def spend_money(self):
        print('You donated some to the people in need. Ate some delicious food which you\'ve never tasted before.')
        print('Went for a loooooong drive on your Rolls Royce.')
        print('Enjoyed with friends.')
        print('Slept with your loving partner in your magnificent mansion.')

human = Happiness()
human.earn_money()
human.spend_money()
```
