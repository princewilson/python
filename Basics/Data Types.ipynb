# Data Types in Python

Python has the following data types *built-in* by default:

- Text Type: `str`
- Numeric Types: `int`, `float`, `complex`
- Sequence Types: `list`, `tuple`,`range`
- Mapping Type: `dict`
- Set Types: `set`,`frozenset`
- Boolean Type: `bool`
- Binary Types: `bytes`,`bytearray`,`memoryview`
