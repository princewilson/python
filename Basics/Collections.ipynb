## Strings:

    - A collection of characters is known as a string. 
    - A string must always be enclosed in single or double quotes.

```python
fname = "Prince"
lname = ' Wilson'
name = fname + lname
print(f'My name is {name}')
```

    - Therefore, two strings can be concatenated using '+' operator

```python
name = 'Prince Wilson'
for i in name:
    print(i, end = ' ')
```

    - We can access and iterate through each character in a string using for loop

```python
first_name = name[0:6]
print(first_name)
print('name[0:] is {}'.format(name[0:]))
print('name[:6] is {}'.format(name[:6]))
print('name[-1:] is {}'.format(name[-1:]))
print('name[:-1] is {}'.format(name[:-1]))
print('name[10:-1] is {}'.format(name[10:-1]))
```

    - We can access a substring in a string using the slice operator
    - We can move from the end of a string using negative index

### String Methods:

### capitalize():
    - Makes the first character uppercase and the rest lower case.

```python
a = 'this Is a sEntence.'
print(a.capitalize())
```

### center():
    - Places the string in the center.
    - Takes two parameters:
        - First is the length of string
        - With what character padding must take place. Default is a space.
        

```python
a = 'this Is a sEntence.'
print(a.center(30, '@'))
```

### count():
    - Returns the number of occurences of a particular substring in a string.

```python
a = 'this Is a sEntence.'
print('The count of {} in {} is {}'.format(a[3:4], a, a.count(a[3:4])))
```

### endswith():
    - Return true if the given string endswith the suffix in the parameter.

```python
a = 'this Is a sEntence.'
print('{} ends with \'tence.\' : {}'.format(a, a.endswith('tence.')))
```

### find():
    - Finds the substring in the given string and return the its first index.
    - The starting index and the ending index of the string can be given as optional arguments.

```python
a = 'this Is a sEntence.'
print('"tence." location in {} is : {}'.format(a, a.find('tence.')))
print('"tence." location in {} from index 10 is : {}'.format(a, a.find('tence.', 10)))
```

### index():
    - Returns the index of the substring in the given string.
    - Raises ValueError when substring is not found.

```python
a = 'this Is a sEntence.'
print('Index of "ten" in {} is : {}'.format(a, a.index('ten')))
```

### isalnum():
    - Returns true if all the characters in the string are alphanumberic and there is atleast one character in the string.
    - Note that space or comma or any other character other than alphabets and numbers will result in false.

```python
a = 'Iwasbornon18thofMay2002'
print('a is alpha numberic? : {}'.format(a.isalnum()))
```

### isalpha():

    - Returns true if all the characters are alphabetic, false otherwise.

```python
a = 'Prince'
print('"{}" is alphabetic? : {}'.format(a, a.isalpha()))
```

### isascii():

    - Returns true if all the characters are ASCII, false otherwise.

```python
a = 'I was born on 18th of May and my favourite ascii character is #.'
print('The sentence "{}" contains all ascii? : {}'.format(a, a.isascii()))
```

### isdigit():

    - Returns true if all the characters in the string are digits, false otherwise.

```python
a = 'Prince1805'
print('Does "{}" contain only digits? : {}'.format(a, a.isdigit()))
```

### isidentifier():

    - Returns true if the string is a valid Python identifier,false otherwise.

```python
a = '12'
b = 'name'
print('"{}" is a valid Python identifier? : {}'.format(a, a.isidentifier()))
print('"{}" is a valid Python identifier? : {}'.format(b, b.isidentifier()))
```

### islower():

    - Returns true if the string is a lowercase string, false otherwise.

```python
a = 'Prince'
b = 'prince'
print('"{}" is a lower case string? : {}'.format(a, a.islower()))
print('"{}" is a lower case string? : {}'.format(b, b.islower()))
```

### isspace():

    - Returns true if the string is a whitespace string, false otherwise.

```python
a = '     '
b = '    p    '
print('a contains only spaces? : {}'.format(a.isspace()))
print('b contains only spaces? : {}'.format(b.isspace()))
```

### istitle():

    - Returns true only if the string contains first letter of each word as upper case and the rest lower, otherwise false.

```python
a = 'Prince Wilson'
b = 'Prince wilson'
print('a is title-cased? : {}'.format(a.istitle()))
print('b is title-cased? : {}'.format(b.istitle()))
```

### isupper():

        - Returns true if the string is upper case, false otherwise.

```python
a = 'UPPER CASE'
print('a is in uppercase? : {}'.format(a.isupper()))
```

### join():

    - The string calling this funtion will be placed in between each character in the parameter string and returns the new string.

```python
a = 'ALIET'
print('Placing dot in between each alphabet in a: {}'.format('.'.join(a)))
```

### lower():

    - Returns a copy of the string converted to lower case.

```python
a = 'SMART'
print('a in upper case: {}'.format(a.lower()))
```

### lstrip():

    - Returns a copy of the string with leading white spaces removed.

```python
a = '       Leading white space.'
print('a : {}'.format(a.lstrip()))
```

### partition():

    - Parititions the string into three parts using the given seperator.
    - If the seperator is found the return value will be a 3-tuple with the first part being the part before the seprator, second part is the seperator itself and third is the rest of the string.
    - If the seperator is not found then the return value will be a 3-tuple with the first element being the original string and the rest two null values.

```python
name = 'Your Prince'
city = 'Vijayawada'
print('name after calling partition() : {}'.format(name.partition(' ')))
print('city after calling partition() : {}'.format(city.partition(' ')))
```

### replace():

    - Returns a copy of all the occurences of the substring replaced with a new one.
    - The first parameter is the substring to be replace and the second is with what to replace.
    - The third parameter is optional where you can specify for how many occurences the substring must be repalced.

```python
a = 'Wheat grinds'
a = a.replace('Wh','Gr')
a = a.replace('gr', 'M')
print('We all have {}'.format(a))
```

### split():

    - Returns a list of strings after seperating the given string with the parameter given.

```python
a = 'Style is the reflection of your attitude'
b = a.split(' ')
print(b)
```

### splitlines():

    - Returns a list of strings based on the lines.

```python
a = 'This is the first line\nThis is the second line\nAnd this is the third'
b = a.splitlines()
print(b)
```

### swapcase():

    - Converts lowercase to uppercase and uppercase to lowercase.

```python
a = 'sWaP tHiS cAsE'
print(a.swapcase())
```

### title():

    - Returns a version of the string where each word is titlecased.

```python
a = 'TiTlE tHiS cAsE'
print(a.title())
```

### upper():

    - Converts a string to uppercase.

```python
a = 'convert this to upper case.'
print(a.upper())
```

## Lists:

    - Lists is a data structure which can used to store a collection of different types of data.
    - Like in strings, you can use the slice operator(:) to access the elements.
    - The syntax for declaring a list type variable is:

`variable_name = [item 1, item 2, item 3, ....]`

```python
names = ['Elon Musk', 'Bill Gates', 'Steve Jobs', 'Sundar Pichai']
print(names[0:2])
```

<p style = "font-family: 'Courier New';">- You can perform various operations by iterating through the list using for loop</p>


```python
fruits = ['apple', 'orange', 'goa', 'mango', 'custard']
print('The names of the fruits which I like are: ')
for i in fruits:
    print(i)
```

### List Methods:

### append():

    - Used to add elements to the list.
    - You can only add one element at a time.

```python
fruits = ['apple', 'orange', 'goa', 'mango', 'custard']
fruits.append('banana')
print(fruits)
```

### clear():

    - Removes all the items from the list.

```python
fruits = ['apple', 'orange', 'goa', 'mango', 'custard']
print(fruits)
fruits.clear()
print(fruits)
```

### count():

    - Counts the number of occurences of a particular value.

```python
nums = [1, 2, 3, 2, 5, 4, 3, 2, 1, 7]
print('Number of occurences of 2 is : {}'.format(nums.count(2)))
print('Number of occurences of 0 is : {}'.format(nums.count(0)))
```

### extend():

    - Append multiple elements to the list from the given iterable.

```python
nums = [1, 2, 3, 4, 5]
print(nums)
nums.extend([6, 7, 8, 9, 10])
print(nums)
```

### index():

    - Returns the first index of the value.

```python
nums = [1, 2, 3, 2, 5, 4, 3, 2, 1, 7]
print('First index of 2 in nums is : {}'.format(nums.index(2)))
```

### insert():

    - To insert a value at a particular index.
    - insert() takes two parameters: the index where you want to insert & the value you want to insert.

```python
nums = [1, 2, 3, 2, 5, 4, 3, 2, 1, 7]
print(nums)
nums.insert(4, 3) # 3 at index 4
print(nums)
```

### pop():

    - Removes the last element of the list.

```python
places = ['California', 'Las Vegas', 'New York']
print(places)
print(places.pop())
print(places)
```

### remove():

    - Takes a value as the parameter and removes that value from the list.

```python
site_seeing = ['Araku', 'Munnar', 'Idukki', 'Palakayam Thattu']
print(site_seeing)
site_seeing.remove('Munnar')
print(site_seeing)
```

### reverse():

    - Reverses the order of the elements in the list.

```python
site_seeing = ['Araku', 'Munnar', 'Idukki', 'Palakayam Thattu']
print(site_seeing)
site_seeing.reverse()
print(site_seeing)
```

### sort():

    - Sorts the element in ascending order and returns null.

```python
nums = [1, 5,3 , 7, 6, 4, 9, 12, 23]
print(nums)
nums.sort()
print(nums)
```

## Taking User Input:

    - We can take the values of the list from the user using for loop and perform our desired operations.
    - Let's write a program to take some integers from the user and print all the even numbers.

```python
nums = []
n = int(input('Enter the number of values to be present in the list: '))
for i in range(n):
    nums.append(int(input()))
print('The even numbers are: ')
for i in nums:
    if i % 2 == 0:
        print(i)
```

## List Comprehensions:

    - We can simplify the initialization of the list using list comprehensions.
    - Below is the program to have a set of even numbers in the list:

```python
n = [i for i in range(11) if i % 2 == 0]
print(n)
```

## Ternary Operator

    - The syntax of the ternary operator is:

`expression if condition else expression` <br>
       
       Which simply means: do this if the condition is true or else do that.
    - Suppose you want to have even numbers in one list and the odd number in another then:

```python
evens = []
odds = []
for i in range(12):
    evens.append(i) if i % 2 == 0 else odds.append(i)
print(evens)
print(odds)
```

## Tuples:

    - Tuple works just like a list but with less flexibility.
    - You cannot add more elements in a tuple
    - You only have 2 methods in a tuple
    - The added advantage of these restrictions is that, it makes iterating through elements faster. So if you have a fixed set of elements, then it's advisable to use tuple.
    - The syntax of a tuple is similar to a list. The only difference is you have parenthesis instead of square brackets.
    - You can access the elements using index numbers.
    - You can also use the slice operator in tuple.    
    - Tuple does not allow comprehensions.

### Tuple Methods:

### count():
    - Returns the frequency of a value in the tuple
    - Takes the value as the parameter.

```python
tup = ('a', 'b', 'c', 'a', 'd', 'e', 'd')
print('The count of a in tup is : {}'.format(tup.count('a')))
```

### index():
    - Returns the first index of a particular value.
    - Takes the value as the parameter.

```python
tup = ('a', 'b', 'c', 'a', 'd', 'e', 'd')
print('The first index of d in tup is : {}'.format(tup.index('d')))
```
